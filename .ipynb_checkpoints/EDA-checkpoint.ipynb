# Data reading

```python
consumer_price_indices = pd.read_csv("./processed data/Consumer_price_indices_for_food_and_services.csv", sep=";")
general_education_schools = pd.read_csv("./processed data/General_education_schools.csv", sep=";")
population = pd.read_csv("./processed data/population.csv", sep=";")
preschool_educational_institutions = pd.read_csv("./processed data/Preschool_educational_institutions.csv", sep=";")
vocational_schools = pd.read_csv("./processed data/Vocational_schools.csv", sep=";")
```

# Primary preprocessing

Firstly, we must present all numeric variables from string to numeric type.

```python
for col in consumer_price_indices.columns[1:]:
    for i in consumer_price_indices.index:
        consumer_price_indices[col][i] = float(consumer_price_indices[col][i].replace(',', '.'))

consumer_price_indices = consumer_price_indices.astype('float')
```

```python
for i in general_education_schools.index:
    general_education_schools["Total_institutions_number"][i] = float(general_education_schools["Total_institutions_number"][i].replace(',', '.'))

for col in general_education_schools.columns[1:]:
    general_education_schools[col] = general_education_schools[col].astype("float")
```

```python
for col in population.columns[1:]:
    for i in population.index:
        population[col][i] = float(population[col][i].replace(',', '.'))

population = population.astype('float')
```

```python
for col in preschool_educational_institutions.columns[1:]:
    for i in population.index:
        preschool_educational_institutions[col][i] = float(preschool_educational_institutions[col][i].replace(',', '.'))

preschool_educational_institutions = preschool_educational_institutions.astype('float')
```

```python
for col in vocational_schools.columns[1:]:
    for i in vocational_schools.index:
        vocational_schools[col][i] = float(vocational_schools[col][i].replace(',', '.'))

vocational_schools = vocational_schools.astype('float')
```

Now let make all data to one format. To do it we must preprocess data about general education schools and make all datasets starting from same year.

```python
general_education_schools
```

```python
general_education_schools["Year"] = [float(yyyy.split("_")[0]) for yyyy in general_education_schools["Year"].to_list()]
general_education_schools
```

```python
general_education_schools = general_education_schools.astype("float")
```

```python
consumer_price_indices.head()
```

```python
general_education_schools = general_education_schools.take(general_education_schools.index[11:])
general_education_schools.index = range(len(general_education_schools))

population = population.take(population.index[11:])
population.index = range(len(population))

preschool_educational_institutions = preschool_educational_institutions.take(preschool_educational_institutions.index[11:])
preschool_educational_institutions.index = range(len(preschool_educational_institutions))

vocational_schools = vocational_schools.take(vocational_schools.index[11:])
vocational_schools.index = range(len(vocational_schools))
```

Now let make one dataset from all having.

```python
data = consumer_price_indices.copy()
data = data.merge(general_education_schools, on="Year")
data = data.merge(population, on="Year")
data = data.merge(preschool_educational_institutions, on="Year")
data = data.merge(vocational_schools, on="Year")
```

```python
data["Year"] = data["Year"].astype("int")
data.info()
```

# Statistical analysis and data cleaning

Ok, we have alone dataset with only numeric parameters. Before building prediction models we must do some caculations to understand behaviour of data.

Firstly, let try to find and filter well-correlated parameters.

```python
well_correlated = []

for i in range(len(data.columns)-1):
    for j in range(i+1, len(data.columns)):
        if abs( data[data.columns[i]].corr(data[data.columns[j]]) )>0.9:
            well_correlated.append([data.columns[i], data.columns[j], data[data.columns[i]].corr(data[data.columns[j]])])

print("Well-correlated parameters:")
for pair in well_correlated:
    print(f"{pair[0]} and {pair[1]}, r={pair[2]}")
```

```python
print(f"General number of well-correlated parameters {len(well_correlated)}")
```

What we may see using this list?
* We may see, that year as parameter we may drop from dataset.
* Also Total_teachers parameter (what we will predict later) is well-correlated with many other parameters.
* We may drop Admitted_students_thsds, Total_students_thsds, Total_present_population_thnds and Total_resident_population_thnds parameters.

Let drop marked parameters from dataset.

```python
cols_to_drop = [
    "Year",
    "Consumer_price_indices",
    "Total_institutions_number",
    "full-time_students",
    "part_time_students_including_distance_learning",
    "complete_secondary_education_diplomas_number",
    "Total_schools_number",
    "Admitted_students_thsds",
    "Total_present_population_thnds",
    "Total_resident_population_thnds",
    "Total_students_thsds",
    "Clothing_and_footwear",
    "Miscellaneous_goods_and_services",
    "incomplete_secondary_education_dipomas_number",
    "Furnishings_household_equipment_and_routine_maintenance_of_the_house",
    "Graduated_thsds"
]

data = data.drop(cols_to_drop, axis="columns")
```

```python
well_correlated = []

for i in range(len(data.columns)-1):
    for j in range(i+1, len(data.columns)):
        if abs( data[data.columns[i]].corr(data[data.columns[j]]) )>0.9:
            well_correlated.append([data.columns[i], data.columns[j], data[data.columns[i]].corr(data[data.columns[j]])])

print("Well-correlated parameters:")
for pair in well_correlated:
    print(f"{pair[0]} and {pair[1]}, r={pair[2]}")
```

```python
print(f"General number of well-correlated parameters {len(well_correlated)}")
```

Now let's select parameter, we will predict, from dataset and calculate some t-tests.

```python
y = data["Total_teachers"] # parameter we will predict
X = data.drop("Total_teachers", axis="columns") # parameters we will use for prediction
```

```python
print("Result parameters list:")
for col in X.columns:
    print(f" - {col}")
```

```python
for col in X.columns:
    result = stats.ttest_rel(y, X[col])
    print(f"For Total_teachers and {col}:\n p-value={result[1]}\n statistics={result[0]}\n")
```

We see, that with =0.01 we must use all selected parameters for prediction.

**Now let's make some visualizations to select correct model.**

```python
print(f"General parameners number: {len(X.columns)}")
```

```python
fig, axes = plt.subplots(7, 2)
fig.set_size_inches(8, 28)

for i in range(7):
    for j in range(2):
        axes[i, j].plot(X[ X.columns[i*2+j] ], y, ".")
        axes[i, j].set_xlabel(X.columns[i*2+j])
        axes[i, j].set_ylabel("Teachers number")
        axes[i, j].grid()

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

We may see, that there aren't any simple structure to build simple prediction model. So, we must experiment with different prediction models to make correct prediction.

# Prediction model selection

Now let's select model to predict total teachers number in the next year.

**Firstly, let's try to answer next questions:**
* May we use linear regression as prediction model in context of using data?
* If no, may we use polynomial regression in this context?

Firstly, let's build and test linear and polynomial repgression models.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
```

```python
linear_regrression = LinearRegression().fit(X, y)

print(f"Max error: {metrics.max_error(y, linear_regrression.predict(X))}")
print(f"Mean absolute error: {metrics.mean_absolute_error(y, linear_regrression.predict(X))}")
print(f"Mean squared error: {metrics.mean_squared_error(y, linear_regrression.predict(X))}")
```

We may see, that linear regression model is excellent! But what if we build quadratic regression model and see values of used errors?

```python
new_X = PolynomialFeatures(degree=2).fit_transform(X)

quadratic_regrression = LinearRegression().fit(new_X, y)

print(f"Max error: {metrics.max_error(y, quadratic_regrression.predict(new_X))}")
print(f"Mean absolute error: {metrics.mean_absolute_error(y, quadratic_regrression.predict(new_X))}")
print(f"Mean squared error: {metrics.mean_squared_error(y, quadratic_regrression.predict(new_X))}")
```

**As we see, that to predict total teachers number in next year it may enough to use linear regression model. Seeing the structure of using data, we may try use linear regression to predict total teachers number in 2024-2027 years.** Let's try it. But let's remember, to prediction we formally use time serieses.

```python
X_for_2023 = X.copy()
y_for_2023 = y.copy()

X_for_2024 = X.take(X.index[:-1])
y_for_2024 = y.take(y.index[1:])

X_for_2025 = X.take(X.index[:-2])
y_for_2025 = y.take(y.index[2:])

X_for_2026 = X.take(X.index[:-3])
y_for_2026 = y.take(y.index[3:])

X_for_2027 = X.take(X.index[:-4])
y_for_2027 = y.take(y.index[4:])
```

```python
model_for_2023 = LinearRegression().fit(X_for_2023, y_for_2023)
model_for_2024 = LinearRegression().fit(X_for_2024, y_for_2024)
model_for_2025 = LinearRegression().fit(X_for_2025, y_for_2025)
model_for_2026 = LinearRegression().fit(X_for_2026, y_for_2026)
model_for_2027 = LinearRegression().fit(X_for_2027, y_for_2027)
```

**Calculating and visualizing of prediction**

```python
y_2023 = model_for_2023.predict(pd.DataFrame(X_for_2023.iloc[21]).T)
y_2024 = model_for_2024.predict(pd.DataFrame(X_for_2024.iloc[20]).T)
y_2025 = model_for_2025.predict(pd.DataFrame(X_for_2025.iloc[19]).T)
y_2026 = model_for_2026.predict(pd.DataFrame(X_for_2026.iloc[18]).T)
y_2027 = model_for_2027.predict(pd.DataFrame(X_for_2027.iloc[17]).T)
```

```python
prediction = [[i for i in range(2023, 2027+1)]]
prediction.append([ 
    y_2023[0],
    y_2024[0],
    y_2025[0],
    y_2026[0],
    y_2027[0]
])

prediction
```

```python
plt.plot(range(2001, 2022+1), data["Total_teachers"], label="Actual teachers number in general schools by year")
plt.plot([2022]+prediction[0], [data.loc[21]["Total_teachers"]]+prediction[1], label="Prediction (using Linear Regression)")
plt.grid()
plt.legend()
plt.xlabel("Year")
plt.ylabel("Total teachers number, thsds")
plt.show()
```

We see, that linear regression gives no very realistic prediction. And it's logical using previous calculations. So, **what we really use regularization?**.

```python
from sklearn.linear_model import Lasso # for L1 regularization
from sklearn.linear_model import Ridge # for L2 regularization
from sklearn.linear_model import ElasticNet # for L1+L2 regularization
```

```python
lasso_regression = Lasso(alpha=0.15).fit(X, y)

print("Errors for Lasso")
print(f"Max error: {metrics.max_error(y, lasso_regression.predict(X))}")
print(f"Mean absolute error: {metrics.mean_absolute_error(y, lasso_regression.predict(X))}")
print(f"Mean squared error: {metrics.mean_squared_error(y, lasso_regression.predict(X))}")
```

```python
ridge_regression = Ridge(alpha=0.15).fit(X, y)

print("Errors for Ridge")
print(f"Max error: {metrics.max_error(y, ridge_regression.predict(X))}")
print(f"Mean absolute error: {metrics.mean_absolute_error(y, ridge_regression.predict(X))}")
print(f"Mean squared error: {metrics.mean_squared_error(y, ridge_regression.predict(X))}")
```

```python
elasticnet_regression = ElasticNet(alpha=0.15).fit(X, y)

print("Errors for Ridge")
print(f"Max error: {metrics.max_error(y, elasticnet_regression.predict(X))}")
print(f"Mean absolute error: {metrics.mean_absolute_error(y, elasticnet_regression.predict(X))}")
print(f"Mean squared error: {metrics.mean_squared_error(y, elasticnet_regression.predict(X))}")
```

Let's visualize predictions, what will be built using these models.

```python
lasso_for_2023 = Lasso(alpha=0.05).fit(X_for_2023, y_for_2023)
lasso_for_2024 = Lasso(alpha=0.05).fit(X_for_2024, y_for_2024)
lasso_for_2025 = Lasso(alpha=0.05).fit(X_for_2025, y_for_2025)
lasso_for_2026 = Lasso(alpha=0.05).fit(X_for_2026, y_for_2026)
lasso_for_2027 = Lasso(alpha=0.05).fit(X_for_2027, y_for_2027)

ridge_for_2023 = Ridge(alpha=0.05).fit(X_for_2023, y_for_2023)
ridge_for_2024 = Ridge(alpha=0.05).fit(X_for_2024, y_for_2024)
ridge_for_2025 = Ridge(alpha=0.05).fit(X_for_2025, y_for_2025)
ridge_for_2026 = Ridge(alpha=0.05).fit(X_for_2026, y_for_2026)
ridge_for_2027 = Ridge(alpha=0.05).fit(X_for_2027, y_for_2027)

elastic_for_2023 = ElasticNet(alpha=0.05).fit(X_for_2023, y_for_2023)
elastic_for_2024 = ElasticNet(alpha=0.05).fit(X_for_2024, y_for_2024)
elastic_for_2025 = ElasticNet(alpha=0.05).fit(X_for_2025, y_for_2025)
elastic_for_2026 = ElasticNet(alpha=0.05).fit(X_for_2026, y_for_2026)
elastic_for_2027 = ElasticNet(alpha=0.05).fit(X_for_2027, y_for_2027)
```

```python
# lasso reguralization
lasso_y_2023 = lasso_for_2023.predict(pd.DataFrame(X_for_2023.iloc[21]).T)
lasso_y_2024 = lasso_for_2024.predict(pd.DataFrame(X_for_2024.iloc[20]).T)
lasso_y_2025 = lasso_for_2025.predict(pd.DataFrame(X_for_2025.iloc[19]).T)
lasso_y_2026 = lasso_for_2026.predict(pd.DataFrame(X_for_2026.iloc[18]).T)
lasso_y_2027 = lasso_for_2027.predict(pd.DataFrame(X_for_2027.iloc[17]).T)

lasso_prediction = [[i for i in range(2023, 2027+1)]]
lasso_prediction.append([ 
    lasso_y_2023[0],
    lasso_y_2024[0],
    lasso_y_2025[0],
    lasso_y_2026[0],
    lasso_y_2027[0]
])

# Ridge reguarization
ridge_y_2023 = ridge_for_2023.predict(pd.DataFrame(X_for_2023.iloc[21]).T)
ridge_y_2024 = ridge_for_2024.predict(pd.DataFrame(X_for_2024.iloc[20]).T)
ridge_y_2025 = ridge_for_2025.predict(pd.DataFrame(X_for_2025.iloc[19]).T)
ridge_y_2026 = ridge_for_2026.predict(pd.DataFrame(X_for_2026.iloc[18]).T)
ridge_y_2027 = ridge_for_2027.predict(pd.DataFrame(X_for_2027.iloc[17]).T)

ridge_prediction = [[i for i in range(2023, 2027+1)]]
ridge_prediction.append([ 
    ridge_y_2023[0],
    ridge_y_2024[0],
    ridge_y_2025[0],
    ridge_y_2026[0],
    ridge_y_2027[0]
])

# ElasticNet regularization
elastic_y_2023 = elastic_for_2023.predict(pd.DataFrame(X_for_2023.iloc[21]).T)
elastic_y_2024 = elastic_for_2024.predict(pd.DataFrame(X_for_2024.iloc[20]).T)
elastic_y_2025 = elastic_for_2025.predict(pd.DataFrame(X_for_2025.iloc[19]).T)
elastic_y_2026 = elastic_for_2026.predict(pd.DataFrame(X_for_2026.iloc[18]).T)
elastic_y_2027 = elastic_for_2027.predict(pd.DataFrame(X_for_2027.iloc[17]).T)

elastic_prediction = [[i for i in range(2023, 2027+1)]]
elastic_prediction.append([ 
    elastic_y_2023[0],
    elastic_y_2024[0],
    elastic_y_2025[0],
    elastic_y_2026[0],
    elastic_y_2027[0]
])
```

```python
plt.plot(range(2001, 2022+1), data["Total_teachers"], label="Actual teachers number in general schools by year")
plt.plot([2022]+lasso_prediction[0], [data.loc[21]["Total_teachers"]]+lasso_prediction[1], ".-", label="Prediction (using Lasso)")

plt.plot(range(2001, 2022+1), data["Total_teachers"], label="Actual teachers number in general schools by year")
plt.plot([2022]+ridge_prediction[0], [data.loc[21]["Total_teachers"]]+ridge_prediction[1], '--', label="Prediction (using Ridge)")

plt.plot(range(2001, 2022+1), data["Total_teachers"], label="Actual teachers number in general schools by year")
plt.plot([2022]+elastic_prediction[0], [data.loc[21]["Total_teachers"]]+elastic_prediction[1], "-.", label="Prediction (using ElasticNet)")

plt.grid()
plt.legend()
plt.xlabel("Year")
plt.ylabel("Total teachers number, thsds")
plt.show()
```

We see, that usage of regularization in case of linear regression is not bad idea, but it gives similar prediction. So, **if we want to have more realistic prediction, we must use prediction model of other type**.
